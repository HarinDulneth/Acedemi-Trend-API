# 01 Data Understanding

## Introduction
This notebook explores the raw datasets for the course enrollment trend prediction system. The goal is to understand the structure, content, and quality of the data, including enrollment numbers, application data, university rankings, job market demand, and international education trends for 57 courses from 2005 to 2023. District-level data is excluded as per the user's request.

## Objectives
- Load and inspect each dataset.
- Summarize data types, shapes, and statistics.
- Identify missing values and potential inconsistencies.
- Visualize key distributions.


## Import Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)
```

## Load Datasets


```python
# Load datasets
enrollments = pd.read_csv('C:/Users/admin/Downloads/course_enrollment_prediction/data/raw/enrollments.csv')
app_2005_2015 = pd.read_csv('C:/Users/admin/Downloads/course_enrollment_prediction/data/raw/Application_2005-2015.csv')
app_2016_2023 = pd.read_csv('C:/Users/admin/Downloads/course_enrollment_prediction/data/raw/Application_2016-2023.csv')
rankings = pd.read_csv('C:/Users/admin/Downloads/course_enrollment_prediction/data/raw/rank_by_university.csv')
job_market = pd.read_csv('C:/Users/admin/Downloads/course_enrollment_prediction/data/raw/job_market_demand_by_field.csv')
international = pd.read_csv('C:/Users/admin/Downloads/course_enrollment_prediction/data/raw/international_trend.csv')
```

## Inspect Datasets


```python
# Function to summarize dataset
def summarize_dataset(df, name):
    print(f'\n\n=== {name} ===')
    print('Shape:', df.shape)
    print('\nColumns:', df.columns.tolist())
    print('\nData Types:\n', df.dtypes)
    print('\nMissing Values:\n', df.isna().sum())
    print('\nFirst 5 Rows:\n', df.head())

# Summarize each dataset
summarize_dataset(enrollments, 'Enrollments')
summarize_dataset(app_2005_2015, 'Applications 2005-2015')
summarize_dataset(app_2016_2023, 'Applications 2016-2023')
summarize_dataset(rankings, 'University Rankings')
summarize_dataset(job_market, 'Job Market Demand')
summarize_dataset(international, 'International Trends')
```

## Visualize Key Distributions


```python
# Plot histogram of enrollments
plt.figure(figsize=(10, 6))
enrollments['enrollments'].hist(bins=30)
plt.title('Distribution of Enrollment Numbers')
plt.xlabel('Enrollments')
plt.ylabel('Frequency')
plt.show()

# Plot histogram of applications (2005-2015)
plt.figure(figsize=(10, 6))
app_2005_2015['applications'].hist(bins=30)
plt.title('Distribution of Applications (2005-2015)')
plt.xlabel('Applications')
plt.ylabel('Frequency')
plt.show()

# Plot histogram of applications (2005-2015)
plt.figure(figsize=(10, 6))
app_2016_2023['applications'].hist(bins=30)
plt.title('Distribution of Applications (2016-2023)')
plt.xlabel('Applications')
plt.ylabel('Frequency')
plt.show()
```
